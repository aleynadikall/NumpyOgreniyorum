### Lineplot (Çizgi Grafiğin) Oluşturulması

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
df = fmri.copy()
df.head()
```

```python
sns.lineplot(x = "timepoint", y = "signal", data = df);
```

#### Çaprazlamalar

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df);
```

```python
# stil ekleme
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", data = df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "event", 
             style = "event", 
             markers = True,
             dashes = False,
             data = df);
#ortalama noktaları işaretlenmiş oldu.
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "region", 
             style = "event", 
             data = df);
```
