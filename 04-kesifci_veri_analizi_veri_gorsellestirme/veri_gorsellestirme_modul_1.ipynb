## **Workshop**

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 4, 9, 16, 25, 36, 49, 64]

x2 = [1, 2, 3, 4, 5, 6, 7, 8]
y2 = [1, 8, 27, 64, 125, 216, 343, 512]
```

```python
plt.plot(x, y)

plt.title("Line Plot")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid()
plt.show()
```

```python
# bir grafik üzerinde iki verisetini de göstermiş olduk.
plt.plot(x, y)
plt.plot(x2, y2)

plt.title("Line Plot")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.grid()
plt.show()
```

```python
plt.plot(x, y, label = "x kare", color = "#ffee56")
plt.plot(x2, y2, label = "x küp", color = "green")

plt.title("Line Plot")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.legend()
plt.grid()
plt.show()
```

## **Subplots**

```python
fig = plt.figure(figsize = (18,5))

#subplot farklı garfikleri yan yana koymamaızı sağlar.
first_plot = fig.add_subplot(1,3,1)  #satır, sütun, hangi garfiğe aitse
first_plot.plot(x, y, color = "red")
first_plot.set_title("Line Plot")
first_plot.grid()

second_plot = fig.add_subplot(1,3,2)
second_plot.scatter(x, y, color = "green")
second_plot.set_title("Scatter Plot")

third_plot = fig.add_subplot(1,3,3)
third_plot.bar(x, y, color = "orange")
third_plot.set_title("Bar Plot")

plt.show()
```
