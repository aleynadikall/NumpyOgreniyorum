### Birleştirme (Join) İşlemleri

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 99
```

```python
df2
```

```python
pd.concat([df1,df2])
```

###### 18.satırdaki birleştirme işlemi doğru bir işlem değildi. 19.satırda daha doğru halini join = "inner" ile yaptık. Bu birleştirme işlemini kesişimleri elemanlardan oluşan değişkenlere göre birleştirdi. Farklı veri setlerinden oluşan bilgileri birileştirmeye çalışmak pek mantıklı değildir. 

###### Bu birleştirme işlemini seçmiş olduğumuz bir deişkenin değerlerine göre de yapabiliriz: join_axes ile yapılıyormuş ancak silinmiş. Daha sonra araştırmayı unutma!!!
